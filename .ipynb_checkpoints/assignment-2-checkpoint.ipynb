```python
import nltk
from nltk import *
import pandas as pd
import numpy as np
from nltk.corpus import words
nltk.download('words')

correct_spellings = words.words()
```

# Part 2 - Spelling Recommender
For this part of the assignment you will create three different spelling recommenders, that each take a list of misspelled words and recommends a correctly spelled word for every word in the list.

For every misspelled word, the recommender should find find the word in correct_spellings that has the shortest distance*, and starts with the same letter as the misspelled word, and return that word as a recommendation.

*Each of the three different recommenders will use a different distance measure (outlined below).

Each of the recommenders should provide recommendations for the three default words provided: ['cormulent', 'incendenece', 'validrate'].

# Jaccard Distance n-grams = 3

```python
def answer_nine(entries=['cormulent', 'incendenece', 'validrate']):

    rec = {}
    recomm = []
    

    for entry in entries:
        min_dist = 25
        for word in correct_spellings:
            if word[0] == entry[0]:
                x = nltk.jaccard_distance(set(nltk.ngrams(word, n=3)),set(nltk.ngrams(entry, n=3)))
                jacc = x
                if jacc < min_dist:
                    min_dist = jacc
                    rec[entry] = word
        recomm.append(rec[entry])
        
                           
#     recommended = [x for x in rec.values()]
#     recs = []
#     recomm = []
#     for entry in entries:
#         x = {(nltk.jaccard_distance(set(nltk.ngrams(entry,n=3)),set(nltk.ngrams(y,n=3)))):y 
#              for y in correct_spellings if y[0] == entry[0]}
#         recs.append(x)
    
    
#     for dictionary in recs:
#         sort = [v for k,v in sorted(dictionary.items(),key=lambda item: item[0])][0]
#         recomm.append(sort)
    
                
    return recomm
answer_nine()
```

# Jaccard Distance, n-grams = 4

```python


def answer_ten(entries=['cormulent', 'incendenece', 'validrate']):
    n = 4
    recs = []
    recomm = []
    for entry in entries:
        x = {(nltk.jaccard_distance(set(nltk.ngrams(entry,n=n)),set(nltk.ngrams(y,n=n)))):y 
             for y in correct_spellings if y[0] == entry[0]}
        recs.append(x)
    
    
    for dictionary in recs:
        sort = [v for k,v in sorted(dictionary.items(),key=lambda item: item[0])][0]
        recomm.append(sort)
    

    
    return recomm
    
answer_ten()
```

# Levenshtein Distance

```python
def answer_eleven(entries=['cormulent', 'incendenece', 'validrate']):
    
    distances = {}
    recommended = []
    
    for entry in entries:
        
        distances[entry] = None
        min_dist = 26
        
        for word in correct_spellings:
            
            if (word[:2] == entry[:2]) and (abs(len(word)-len(entry) < 3)):

                m = len(entry)
                n = len(word)

                matrix = np.zeros((m+1,n+1))

                for i in range(1,m+1): matrix[i][0] = i
                for j in range(1,n+1): matrix[0][j] = j


                for i in range(1,m + 1):
                    for j in range(1,n + 1):

                        if entry[i-1] == word[j-1]:
                            cost = 0                            
                        else:
                            cost = 1

                        matrix[i][j] = min((matrix[i-1][j] + 1),
                                            (matrix[i][j-1] + 1),
                                            (matrix[i-1][j-1] + cost))

                lev_distance = matrix[m][n]
                if lev_distance < min_dist:
                    min_dist = lev_distance
                    distances[entry] = {word:min_dist}
            else:
                continue
    
    for i in distances.values():
        for j in i.keys():
            recommended.append(j)
            
    return recommended
    
answer_eleven()
```
